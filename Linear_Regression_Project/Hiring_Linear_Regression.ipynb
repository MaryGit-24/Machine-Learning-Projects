In exercise folder hiring.csv.

The data contains hiring statics for a firm such as experience of candidate, his written test score and personal interview score. Based on these 3 factors, HR will decide the salary.

Given this data, you need to build a machine learning model for HR department that can help them decide salaries for future candidates.


Using this predict salaries for following candidates,
1. 2 yr experience, 9 test score, 6 interview score
2. 12 yr experience, 10 test score, 10 interview score.

```python
#load data

hiring = pd.read_csv('/content/drive/MyDrive/WTF-Material_Class B/hiring (1).csv')
hiring
```

```python
#creating a numeric column for experience with map (converting strings to float)

experience_map = {
    'five': 5,
    'two': 2,
    'seven': 7,
    'three': 3,
    'ten': 10,
    'eleven': 11

}
hiring['experience'] = hiring['experience'].map(experience_map)
hiring
```

```python
#calculating experience mean to get a middle number for NaN

hiring['experience'].mean()
```

```python
#filling NaN with experience mean

hiring['experience'].fillna(hiring['experience'].mean(), inplace=True)
hiring
```

```python
#converting float to integer type

hiring['experience'] = hiring['experience'].round().astype(int)
hiring
```

```python
#calculating test_score(out of 10) median to get a middle number for NaN

hiring['test_score(out of 10)'].median()
```

```python
#filling NaN with test_score(out of 10) median

hiring['test_score(out of 10)'].fillna(hiring['test_score(out of 10)'].median(), inplace=True)
hiring
```

```python
#splitting dependent variables from independent variables

indpt_var = hiring.drop('salary($)', axis='columns')
indpt_var
```

```python
#creating a dependent variable

dpt_var = hiring['salary($)']
dpt_var
```

```python
#train model with the independent and dependent variables

from sklearn.linear_model import LinearRegression

reg = LinearRegression()
reg.fit(indpt_var, dpt_var)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(indpt_var, dpt_var)
```

```python
# 2 yr experience, 9 test score, 6 interview score

reg.predict([[2, 9, 6]])
```

```python
# 12 yr experience, 10 test score, 10 interview score

reg.predict([[12, 10, 10]])
```
