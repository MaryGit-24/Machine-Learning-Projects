Mulivariate linear regression in machine learning is an extenstion of linear regression whenmutiple independent variables are used to predict a single dependent variable. The relationship between the independent variable and the dependent vairable is Represented by Linear Equation

```python
import pandas as pd
import numpy as np
import csv
import matplotlib.pyplot as plt
%matplotlib inline
from google.colab import drive
drive.mount('/content/drive')
from sklearn import linear_model
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("/content/drive/MyDrive/WTF-Material_Class B/homeprices 2 (2).csv")
df
```

```python
#Data pre-processing: Fill NA values with the median value of a column

df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

Where area, bedrooom, and age are called the independent variable or features

Whereas prices is a dependant variable

```python
#split the variables

new_df = df.drop('price', axis='columns')
new_df
```

```python
price = df['price']
price
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df, price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.score(new_df, price)
```
